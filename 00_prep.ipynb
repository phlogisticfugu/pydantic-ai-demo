```python
import os
from dotenv import load_dotenv
from rich import print as rich_print

load_dotenv()
```

# Convert from HTML to Markdown

```python
from tavily import TavilyClient

tavily_client = TavilyClient(api_key=os.environ['TAVILY_API_KEY'])
```

```python
response = tavily_client.extract('https://en.wikipedia.org/wiki/George_Washington')
```

```python
with open("GEORGE_WASHINGTON.md", "w", encoding="utf-8") as f:
    f.write(response['results'][0]['raw_content'])
```

# Summarize

```python
with open("GEORGE_WASHINGTON.md", "r", encoding="utf-8") as f:
    wikipedia_bio = f.read()
```

```python
import ollama
from pydantic_ai.models.openai import OpenAIChatModel
from pydantic_ai.providers.ollama import OllamaProvider

OLLAMA_MODEL_NAME = "deepseek-r1:7b"
OLLAMA_BASE_URL = "http://localhost:11434"

ollama.pull(OLLAMA_MODEL_NAME)

model = OpenAIChatModel(
    model_name=OLLAMA_MODEL_NAME,
    provider=OllamaProvider(base_url=f"{OLLAMA_BASE_URL}/v1", api_key="ollama"),
)
```

```python
from pydantic_ai import Agent

agent = Agent(model)
```

```python
result = await agent.run(f"""
    Summarize the biography of George Washington. Focus on major accomplishments and demonstrations of skill and leadership.
    Keep details like the years when different events occurred, but dont include things like quotes, citations, or other irrelevant text.
    Highlight accomplishments, military strategies, presidential policies, and other SPECIFIC details.

    biography:
    {wikipedia_bio}
    """)
summary = result.output
rich_print(result.usage())
```

```python
with open("GEORGE_WASHINGTON_SUMMARY.md", "w", encoding="utf-8") as f:
    f.write(summary)
```
